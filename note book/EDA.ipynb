```python
columns = df.columns

outlier_features = []

for col in columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    
    iqr = q3-q1

    lower_pound = q1-1.5*iqr
    upper_pound = q3+1.5*iqr
    mask = df[(df[col]<lower_pound) | (df[col]>upper_pound)]
    if not mask.empty:
        outlier_features.append([col, mask.shape[0]])
print(outlier_features)
```

```python
cols = [item[0] for item in outlier_features]
counts = [item[1] for item in outlier_features]

plt.figure(figsize=(12,6))
plt.bar(cols, counts, color='red')
plt.xticks(rotation=90)
plt.title("Number of Outliers per Feature")
plt.ylabel("Count of Outliers")
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, cmap = 'coolwarm',linewidth = 0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
class_1 = df[df.Class == 1]
class_0 = df[df.Class == 0]
class_0_sample = class_0.sample(n=len(class_1) , random_state = 42)
under_sampled_df = pd.concat([class_0_sample,class_1] ,axis = 0)
print(under_sampled_df['Class'].value_counts())
plt.figure(figsize=(12,8))
corr = under_sampled_df.corr()
sns.heatmap(corr, cmap = 'coolwarm',linewidth = 0.5)
plt.title('Under Sampled Correlation Matrix')
plt.show()
```

```python
def show_histograms(df):
    fig, axs = plt.subplots(6,5, figsize=(15,12))
    data = df.copy()
    data.Class = data.Class.map({0 : 'No Fraud', 1 : 'Fraud'})
    
    for i, ax in enumerate(axs.flat):
        if i < len(df.columns):  
            feature = df.columns[i]
            sns.histplot(data=data, x=feature, hue='Class', bins=50, 
                         palette='coolwarm', ax=ax, kde=False, element="step", stat="density")
            ax.set_title(feature)
        else:
            ax.axis('off') 
    
    plt.suptitle('Histogram Distribution of Features Per Class')
    plt.tight_layout()
    plt.show()
show_histograms(df)
```

```python
def show_distributions(df):

    fig,axs = plt.subplots(6,5,figsize=(15,12))
    data = df.copy()
    data.Class = data.Class.map({0 : 'No Fraud' , 1 : 'Fraud'})
    for i,ax in enumerate(axs.flat):
        feature = df.columns[i]
        sns.boxplot(x='Class',y = feature, ax = ax,data = data, palette ='coolwarm', hue='Class')
        
        ax.set_title(feature)
    plt.suptitle('Distribution of Features Per Class')
    plt.tight_layout()
    plt.show()

show_distributions(df)
```

# data cleaning


```python
df.drop_duplicates(inplace=True)
```

```python
columns = ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
        'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
        'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']


for col in columns:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    
    iqr = q3-q1

    lower_pound = q1-1.5*iqr
    upper_pound = q3+1.5*iqr
    df[col] = np.where(df[col]<lower_pound, lower_pound,np.where(df[col]>upper_pound, upper_pound, df[col]))
```

# feature engineering

```python
def feature_engineering(df):
    df = df.copy()
    df['Amount_log'] = np.log1p(df['Amount'])
    df['Hour'] = (df['Time'] // 3600) % 24     
    df['Day'] = (df['Time'] // (3600 * 24))    
    return df

df = feature_engineering(df)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(df['Amount'], bins=50, kde=True, color='red')
plt.title("Original Amount Distribution")

plt.subplot(1,2,2)
sns.histplot(df['Amount_log'], bins=50, kde=True, color='blue')
plt.title("Log-transformed Amount Distribution")

plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Hour', hue='Class', data=df, palette='coolwarm')
plt.title("Fraud vs No Fraud by Hour")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Day', hue='Class', data=df, palette='coolwarm')
plt.title("Fraud vs No Fraud by Day")
plt.show()
```
